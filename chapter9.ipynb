```python
def main(file_name):

#variable for storing string
    string_contents =''
    #opens file
    with open(file_name, "r") as file:
        #Stores all data as one string
        string_contents = file.read()
    #calls the list function
    word_list = wordcount.get_words_list(string_contents)
     #calls the count function and gets all the words in the list
    word_counts = wordcount.get_counts(word_list)
 
    for word, count in word_counts.items():
        print(f"{word}: {count}")
  
#calls the file with the text in it
main("some_text.txt")
        
```

```python
Start

Input the text
    ↓
Split the text into words
    ↓
Initialize an empty dictionary (word_count)
    ↓
For each word in the list of words:
    ↓
Check if the word is already a key in word_count:
        ↓
If yes, increment its count by 1
        ↓
If no, add it to word_count with count 1
    ↓
Output the word_count dictionary
    ↓
End
```
